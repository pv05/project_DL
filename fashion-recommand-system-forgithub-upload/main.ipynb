```python
a = round((final_df.isnull().mean()*100),2)
mssingcolumn=[]
missingdata=[] 
missingnum=[]
for i in range(len(a)):
    if a.values[i] != 0:
        mssingcolumn.append(a.keys()[i])
for i in mssingcolumn:
    missingdata.append(round((final_df[i].isnull().mean()*100),2))
    missingnum.append(final_df[i].isnull().sum())
missing=pd.DataFrame({'column':mssingcolumn,'missing percent':missingdata,'missing num':missingnum}) 
missing
```
